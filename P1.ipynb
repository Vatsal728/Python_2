```python
df = df.dropna(subset=['WinningTeam'])
print(df)
```

```python
df.Team1.nunique()
```

```python
d={"Delhi Daredevils" : "Delhi Capitals" , "Kings XI Punjab" :"Punjab Kings" , "Rising Pune Supergiants" :
   "Pune Warriors" , "Rising Pune Supergiant" : "Pune Warriors" , "Gujarat Titans" : "Gujarat Lions"}
df = df.replace(d.keys(),d.values())
df.Team1.nunique()
```

```python
teams = df.Team1.unique()
L = []
for team in teams:
    played = df[(df.Team1 == team)|(df.Team2 == team)].shape[0]
    win = df[(df.WinningTeam == team)].shape[0]
    Home_played = df[(df.Team1 == team)].shape[0]
    Home_win = df[(df.WinningTeam == team) & (df.Team1 == team)].shape[0]
    Away_played = df[(df.Team2 == team)].shape[0]
    Away_win = df[(df.WinningTeam == team) & (df.Team2 == team)].shape[0]
    L.append([teams,played,(win/played)*100,(Home_win/Home_played)*100 , (Away_win/Away_played)*100])
# print(L)
new_df = pd.DataFrame(L , columns=["team","play","win%","Home%","Away%"])
```

```python
print(new_df)
```

```python
df2022 = df[df.Season=='2022']
L1 = []
teams = df2022.Team1.unique()
for team in teams:
    play = df2022[(df2022.Team1 == team)|(df2022.Team2 == team)].shape[0]
    win = df2022[(df2022.WinningTeam == team)].shape[0]
    L1.append([team , play , win , win*2])
new = pd.DataFrame(L1 , columns=["team","played","win","points"])
```
